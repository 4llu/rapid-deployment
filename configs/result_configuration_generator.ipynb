```python
import os

from ruamel.yaml import YAML

yaml = YAML()
yaml.default_flow_style = None
```

# ARotor Replication Dataset Configurations

The new dataset.

Both mixed & non-mixed use the same configuration overrides (torques, rpms, installations, severities). Only the base configurations change: `result_generation_base.yaml` and `result_generation_mixed_base.yaml`.


```python
rpm_combinations = [
    ([500, 1500], [750], [1250]),
    ([500, 1500], [1250], [750]),
    ([750, 1250], [500], [1500]),
    ([750, 1250], [1500], [500]),
]

torque_combinations = [
    ([6, 11], [6], [1]),
    ([1, 11], [1], [6]),
    ([1, 6], [1], [11]),
]

# ? Should validation use both installations?
installation_combinations = [
    ([1], [1], [1, 2]),
    ([2], [2], [1, 2]),
]

# ? Should validation be same as train or test?
severity_combinations = [
    (["severe"], ["severe"], ["mild", "severe"]),
    (["mild"], ["mild"], ["mild", "severe"]),
]


# Create configurations
i = 0

for rpm_comb in rpm_combinations:
    for tor_comb in torque_combinations:
        for ins_comb in installation_combinations:
            for sev_comb in severity_combinations:

                config_override = {
                    "train_rpm": rpm_comb[0],
                    "validation_rpm": rpm_comb[1],
                    "test_rpm": rpm_comb[2],
                    #
                    "train_torques": tor_comb[0],
                    "validation_torques": tor_comb[1],
                    "test_torques": tor_comb[2],
                    #
                    "train_installations": ins_comb[0],
                    "validation_installations": ins_comb[1],
                    "test_installations": ins_comb[2],
                    #
                    "train_severities": sev_comb[0],
                    "validation_severities": sev_comb[1],
                    "test_severities": sev_comb[2],
                }

                path = os.path.join("arotor_replication", "result_generation", f"config_override_{i}.yaml")
                with open(path, "wb") as stream:
                    yaml.dump(data=config_override, stream=stream)
                i += 1
```
