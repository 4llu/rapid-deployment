```python
import os

import joblib
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
sns.set_context("poster", font_scale=0.9)
plt.rcParams["svg.fonttype"] = "none"

print(sns.color_palette().as_hex())
```

```python
sns.color_palette()
```

# Article


```python
data = joblib.load(
    os.path.join(
        os.pardir,
        "RAW",
        "rapid_classification_results",
        "ARotor_replication_torque1.pkl",
    )
)
print(data["accuracies"].shape)
print(data["cfs"].shape)

print("Accuracy:", data["accuracies"].mean() * 100)

categories = ["Healthy", "Pitting", "Wear", "Micropitting", "TFF"]

# print()
# print(data["cfs"][0].shape)
# print(data["cfs"][1])
# print(data["cfs"][1].sum(axis=0))
# raise

cf = data["cfs"].sum(axis=0)
cf = cf / cf.sum(axis=1) * 100
print("Accuracy 2:", np.diag(cf).mean())  # Verifiation

fig, axs = plt.subplots(figsize=(12, 10))

g = sns.heatmap(
    cf,
    annot=True,
    fmt=".1f",
    linewidth=0.3,
    xticklabels=categories,
    yticklabels=categories,
    vmin=0,
    vmax=100,
    cmap="viridis",
)

axs.xaxis.set_ticks_position("top")
axs.xaxis.set_label_position("top")
plt.savefig("images/RC_AGFD_T1_CF.svg")
```
